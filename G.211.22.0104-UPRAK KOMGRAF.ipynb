B.Atribut Stock Name dan Symbol yang memiliki Dividend / Share 5 tertinggi

```python
import pandas as pd
data = pd.read_csv('datauprakkomgraf.csv')  # menggunakan file CSV
sorted_data = data.sort_values(by='Dividend/share', ascending=False)
top_5_dividend = sorted_data.head(5)[['Stock Name', 'Symbol', 'Dividend/share']]
print(top_5_dividend)
```
